# Step 4. Map the features into numbers

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("../input/filled_train.csv")
test = pd.read_csv("../input/filled_test.csv")
all_data = [train, test]
print("Train shape: ", train.shape)
print("Test shape: ", test.shape)
```

```python
train.head()
```

```python
test.head()
```

I want to convert:
* Name: Into pandas dummies.
* Sex. Into booleans.
* Cabin. Into pandas dummies.
* Embarked. Into pandas dummies.

```python
train = pd.get_dummies(train, columns = ["Name"], prefix="N")
train["Sex"] = train["Sex"].map({"male": 0, "female":1})
train = pd.get_dummies(train, columns = ["Cabin"], prefix="C")
train = pd.get_dummies(train, columns = ["Embarked"], prefix="E")

test = pd.get_dummies(test, columns = ["Name"], prefix="N")
test["Sex"] = test["Sex"].map({"male": 0, "female":1})
test = pd.get_dummies(test, columns = ["Cabin"], prefix="C")
last_column = test["C_X"]
test.drop("C_X", axis = 1, inplace = True)
test["C_T"] = 0
test["C_X"] = last_column
test = pd.get_dummies(test, columns = ["Embarked"], prefix="E")
```
