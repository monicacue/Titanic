Ticket sometimes have letter on sometimes just numbers

```python
Ticket = []
for i in list(train.Ticket):
    if not i.isdigit() :
        Ticket.append(i.replace(".","").replace("/","").strip().split(' ')[0]) #Take prefix
    else:
        Ticket.append("X")
        
train["Tick"] = Ticket
train.head(10)
```

```python
train["Tick"].value_counts()
```

## Export the serie
