## Import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../input/clean_train.csv")
test = pd.read_csv("../input/test.csv")
all_data = [train, test]
```

```python
# Create a family size descriptor from SibSp and Parch
train["Fsize"] = train["SibSp"] + train["Parch"] + 1
test["Fsize"] = test["SibSp"] + test["Parch"] + 1
train.head()
```

```python
test.head()
```

```python
with sns.axes_style("white"):
    g = plt.subplots(figsize=(12, 8))
    g = sns.barplot(x="Fsize", y="Survived", data=train)
    g = g.set_ylabel("Survival Probability")
```

## Export serie

```python
train["Fsize"].to_csv("../input/train_values/family_size.csv", index=False)
```

```python
test["Fsize"].to_csv("../input/test_values/family_size.csv", index=False)
```
